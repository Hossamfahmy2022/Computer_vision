```python
import numpy as np
from matplotlib import pyplot as plt
import cv2
```

```python
img = cv2.imread('lena.jpg')
```

2D Convolution ( Image Filtering )


```python
kernel = np.ones((5, 5), np.float32)/25
dst = cv2.filter2D(img, -1, kernel) 
cv2.imshow('filter2D',dst)
cv2.imshow('img',img)
cv2.waitKey(0)
cv2.destroyAllWindows()    
```

 average

```python
blur = cv2.blur(img, (5, 5)) # average
cv2.imshow('blur',blur)
cv2.imshow('img',img)
cv2.waitKey(0)
cv2.destroyAllWindows()    
```

weighted average (GaussianBlur)

```python
gblur = cv2.GaussianBlur(img, (5, 5), 0) #weighted average
cv2.imshow('weighted average',gblur)
cv2.imshow('img',img)
cv2.waitKey(0)
cv2.destroyAllWindows()   
```

medianBlur

```python
median = cv2.medianBlur(img, 5)
cv2.imshow('median',median)
cv2.imshow('img',img)
cv2.waitKey(0)
cv2.destroyAllWindows()   
```

bilateralFilter

```python
bilateralFilter = cv2.bilateralFilter(img, 9, 75, 75) # focus on edges
cv2.imshow('bilateralFilter',bilateralFilter)
cv2.imshow('img',img)
cv2.waitKey(0)
cv2.destroyAllWindows()   
```

```python
img = cv2.imread('lena.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

kernel = np.ones((5, 5), np.float32)/25
dst = cv2.filter2D(img, -1, kernel) 
blur = cv2.blur(img, (5, 5)) # average
gblur = cv2.GaussianBlur(img, (5, 5), 0) #weighted average
median = cv2.medianBlur(img, 5)
bilateralFilter = cv2.bilateralFilter(img, 9, 75, 75) # focus on edges
titles = ['image', '2D Convolution', 'blur', 'GaussianBlur', 'median', 'bilateralFilter']
images = [img, dst, blur, gblur, median, bilateralFilter]
for i in range(6):
    plt.subplot(2, 3, i+1), plt.imshow(images[i], 'gray')
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])#ticks

plt.show()#show
```
